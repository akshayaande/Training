```python
import pandas as pd
```

```python
df = pd.read_csv("Assests/WA_Fn-UseC_-HR-Employee-Attrition.csv")
df.head() #Display the first 5 rows of the DataFrame
```

```python
list = df.sort_values(by = "YearsAtCompany" , ascending = False)
list.head()
```

```python
df["SeniorEmployees"] = df["Age"] >50
df.info()
df["Attrition_values"] = df["Attrition"].map({"Yes":1,"No" : 0})
print(df[["Attrition" , "Attrition_values"]].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize =(10,6))
sns.countplot(x="Attrition", data=df)
plt.title("Employee Attrition Count")
plt.xlabel("Attrition")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize =(10,6))
sns.histplot(df["Age"],bins = 30 , kde =True,color="Red")
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="Attrition" , y="MonthlyIncome" ,data=df , color = "pink")
plt.title("Monthly Income by Attrition")
plt.xlabel("Attrition")
plt.ylabel("MonthlyIncome")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df ,x="JobSatisfaction" , hue="Attrition")
plt.title("Job Satisfaction by Attrition")
plt.xlabel("Job Satisfaction")
plt.ylabel("Count")
plt.legend(title = "Attrition")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.kdeplot(data=df,x="DistanceFromHome" , hue="Attrition" , fill = True)
plt.title("Distance From  Home by Attrition")
plt.xlabel("Distance From Home")
plt.ylabel("Density")
plt.legend(title = "Attrition")
plt.show()
```
